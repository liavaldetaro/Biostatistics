```python
cancer = pd.read_csv('data/cancer.csv', sep=',', decimal=".")
cancer['log_surv'] = np.log(cancer['surv'])

print(cancer)
```

### 1) Is there a statistical significant difference in the median survival for the five types of cancer?

```python
%matplotlib inline

fig, ax = plt.subplots(1, 1)
ax = sns.boxplot(x='type', y='surv', data=cancer)
ax = sns.swarmplot(x='type', y='surv', data=cancer, color='red')
plt.show()

fig, ax = plt.subplots(1, 1)
ax = sns.boxplot(x='type', y='log_surv', data=cancer)
ax = sns.swarmplot(x='type', y='log_surv', data=cancer, color='red')
plt.show()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(2, 3, 1)
sm.graphics.qqplot(cancer['surv'].loc[cancer['type'] == 'Stomach'], ax=ax1, fit=True, line='45')
ax1.set_title('Stomach')

ax2 = fig.add_subplot(2, 3, 2)
sm.graphics.qqplot(cancer['surv'].loc[cancer['type'] == 'Bronchus'], ax=ax2, fit=True, line='45')
ax2.set_title('Bronchus')

ax3 = fig.add_subplot(2, 3, 3)
sm.graphics.qqplot(cancer['surv'].loc[cancer['type'] == 'Colon'],  ax=ax3, fit=True, line='45')
ax3.set_title('Colon')

ax4 = fig.add_subplot(2, 3, 4)
sm.graphics.qqplot(cancer['surv'].loc[cancer['type'] == 'Ovary'],  ax=ax4, fit=True, line='45')
ax4.set_title('Ovary')

ax5 = fig.add_subplot(2, 3, 5)
sm.graphics.qqplot(cancer['surv'].loc[cancer['type'] == 'Breast'],  ax=ax5, fit=True, line='45')
ax5.set_title('Breast')

plt.tight_layout()
plt.show()
```

The data is not normally distributed -> let's try the log surv

```python
%matplotlib inline
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(2, 3, 1)
sm.graphics.qqplot(cancer['log_surv'].loc[cancer['type'] == 'Stomach'], ax=ax1, fit=True, line='45')
ax1.set_title('Stomach')

ax2 = fig.add_subplot(2, 3, 2)
sm.graphics.qqplot(cancer['log_surv'].loc[cancer['type'] == 'Bronchus'], ax=ax2, fit=True, line='45')
ax2.set_title('Bronchus')

ax3 = fig.add_subplot(2, 3, 3)
sm.graphics.qqplot(cancer['log_surv'].loc[cancer['type'] == 'Colon'],  ax=ax3, fit=True, line='45')
ax3.set_title('Colon')

ax4 = fig.add_subplot(2, 3, 4)
sm.graphics.qqplot(cancer['log_surv'].loc[cancer['type'] == 'Ovary'],  ax=ax4, fit=True, line='45')
ax4.set_title('Ovary')

ax5 = fig.add_subplot(2, 3, 5)
sm.graphics.qqplot(cancer['log_surv'].loc[cancer['type'] == 'Breast'],  ax=ax5, fit=True, line='45')
ax5.set_title('Breast')

plt.tight_layout()
plt.show()
```

We can now do a Bartlett test, to see if the hypothesis of equal variances hold 

```python
print(stats.bartlett(cancer['log_surv'].loc[cancer['type'] == 'Stomach'], 
                    cancer['log_surv'].loc[cancer['type'] == 'Bronchus'],
                    cancer['log_surv'].loc[cancer['type'] == 'Colon'],
                    cancer['log_surv'].loc[cancer['type'] == 'Ovary'],
                   cancer['log_surv'].loc[cancer['type'] == 'Breast'] ))
```

We find a p-value of 0.3. Therefore, we can assume that the standard deviations of the log-transformed survival times are equal. With the additional assumption of independent measurements and that each group follows a normal distribution, we can do the ANOVA analysis:

```python
import pingouin as pg

aov = pg.anova(data=cancer, dv='log_surv', between='type', detailed=True)
print('ANOVA: \n', aov, '\n\n')
```

We found a p-value of <0.005. Therefore, there is a statistically significant difference between the 5 groups' mean log-survival.

 Since we know that an estimate of the mean on the log-scale turns in to an estimate of the median when it is
"back-transformed" (using the exponential function) to the original scale (see day 2), this is
equivalent to saying that there is a statistical significant difference between the median
survival times in the five groups. <- from the answer sheet. I need to check this on day 2 notes...

### 2) Estimate the median survival in breast and stomach cancer.  

```python
import statsmodels.stats.api as sms

print(' ********* Analysis using a one-sample normal analysis ********* \n \n')
print('----------- Breast cancer ---------')
mean_ci = sms.DescrStatsW(cancer['log_surv'].loc[cancer['type'] == 'Breast']).tconfint_mean()
print('Mean (log-scale)', np.mean(cancer['log_surv'].loc[cancer['type'] == 'Breast']), ', 95% CI:', mean_ci, '\n')

print('Applying a back-transformation, we get: \n')
print('Median survival', np.exp(np.mean(cancer['log_surv'].loc[cancer['type'] == 'Breast'])), ', 95% CI:', np.exp(mean_ci), 'days','\n')

print('----------- Stomach cancer ---------')
mean_ci = sms.DescrStatsW(cancer['log_surv'].loc[cancer['type'] == 'Stomach']).tconfint_mean()
print('Mean (log-scale)', np.mean(cancer['log_surv'].loc[cancer['type'] == 'Stomach']), ', 95% CI:', mean_ci, '\n')
a = cancer['log_surv'].loc[cancer['type'] == 'Stomach']

print('Applying a back-transformation, we get: \n')
print('Median survival', np.exp(np.mean(cancer['log_surv'].loc[cancer['type'] == 'Stomach'])), ', 95% CI:', np.exp(mean_ci), 'days','\n')
```

```python
#First, we define a subset of the data set (only including breast and stomach patients)
cancer_subset = cancer.loc[(cancer['type'] == 'Breast') | (cancer['type'] =='Stomach')]
cancer_subset['index'] = 0
cancer_subset['index'][cancer_subset['type'] == 'Breast'] = 1
```

```python
print(' ********* Analysis using a linear regression analysiss ********* \n \n')

X = sm.add_constant(cancer_subset[['index']])
Y = cancer_subset['log_surv']
model = sm.OLS(Y, X).fit()
print(model.summary())

print('----------- Breast cancer ---------')
print('Mean (log-scale)', 4.96, ', 95% CI: ', [4.13, 5.799], '\n')

print('Applying a back-transformation, we get: \n')
print('Median survival', np.exp(4.96), ', 95% CI:', np.exp([4.13, 5.799]), 'days','\n')

print('----------- Stomach cancer ---------')
print('Mean (log-scale)', 6.55, ', 95% CI: ', [4.5, 5.799], '\n')

print('Applying a back-transformation, we get: \n')
print('Median survival', np.exp(6.5586), ', 95% CI:', np.exp([5.655,7.462]), 'days','\n')
```

### 3) Estimate the ratio between the median survival in breast and stomach cancer. 


From the linear regression in the log scale, we got the coefficients from

##### log(time to death in group 2) = log(const) + log(time to death in group 1)

Therefore, the ratio between the median survival for the two groups is simply the intercept value for the line fit: 

```python
print('median survival ratio: \n', np.exp(1.59), ', 95% CI:', np.exp([0.363, 2.818]), 'days')
```

Thus, the median survival time in the breast cancer group is 4.9 times longer than the stomach cancer group.
